# Importing Libraries

```python
import numpy as np                  #numpy library and alias np
import matplotlib.pyplot as plt  #accessing pyplot from matplotlib
import pandas as pd
```

# Importing Dataset

```python
dataset = pd.read_csv("Social_Network_Ads.csv") #accesing read_csv frpm pd lib and reading Data
X = dataset.iloc[:,0:-1].values    #matrix of features , matrix of feature is the matrix of data table where the features is described [2]
y = dataset.iloc[:,-1].values    #matrix of dependables

print(dataset)
```

# Splitting the Data into training and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

#  Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Training Logistic Regression model on training set

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

# predicting new result

```python
purchase_decision=classifier.predict(X_test[0].reshape(1,-1))
if purchase_decision==0:
    print("This user wont purchase")
else:
    print("This user will purchase")
```

# Predicting a random unseen data

```python
purchase_decision=classifier.predict(sc.transform([[24,2500]]))
if purchase_decision==0:
    print("This user wont purchase")
else:
    print("This user will purchase")
```

# Predicting Test set Result

```python
y_pred =classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),axis=1))
```

# Making the Confusion Matrix

```python
from sklearn import metrics
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
confusion_matrix
```

# Displaying Confusion Metrix

```python
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[False,True])
cm_display.plot()
plt.show()
```

# Created Metrics : Accuracy

```python
Accuracy = metrics.accuracy_score(y_test,y_pred)
print(Accuracy)
```

# Created Metrics : Precision

```python
Precision = metrics.precision_score(y_test,y_pred)
print(Precision)
```

# Created Metrics : Sensitivity (Recall)

```python
Recall = metrics.recall_score(y_test,y_pred)
print(Recall)
```

# Created Metrics : Specificity

```python
Specificity = metrics.recall_score(y_test,y_pred,pos_label=0)
print(Specificity)
```

# Created Metrics :  F-score

```python
F1_score = metrics.f1_score(y_test,y_pred)
print(F1_score)
```
